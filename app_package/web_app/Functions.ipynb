The function that gets all csv files into one data frame from data_sources folder

```python
def get_data_frame():
    path = r'..\data_sources'
    filenames = glob.glob(os.path.join(path + "/*.csv"))
    li = []

    for filename in filenames:
        df = pd.read_csv(filename, index_col=None, header=0)
        df['Country'] = filename.split('\\')[2].split('_')[0]
        li.append(df)

    
    frame = pd.concat(li, axis=0, ignore_index=True)

    return frame

df = get_data_frame()
df.head()

# we can also use this as `from utilities import get_data_frame`
```

1. The function that gives country spesific dataframe by using given Data frame and Country name

```python
def filter_by_country(frame, country):
    df = pd.DataFrame(frame)
    df = df[df['Country'] == country]
    return df

aus_data_frame = filter_by_country(df, "Australia")
aus_data_frame.head()
```

2. The function that gives country and region spesific dataframe by using given Data frame, Country name and
       region

```python
def filter_by_country_and_region(frame, country, region):
    df = pd.DataFrame(frame)
    df = df[df['Country'] == country]
    df = df[df['Region'].str.contains(region, na=False)]
    return df

aus_barossa_data_frame = filter_by_country_and_region(df, "Australia", "Barossa")
aus_barossa_data_frame.head()
```

3. The function that gives country, region and wine style spesific dataframe by using given Data frame, Country    
       name, region and wine style

```python
def filter_by_country_and_region_and_wine_style(frame, country, region, wine_style):
    df = pd.DataFrame(frame)
    df = df[df['Country'] == country]
    df = df[df['Region'].str.contains(region, na=False)]
    df = df[df['Wine style'] == wine_style]
    return df

aus_barossa_shiraz_data_frame = filter_by_country_and_region_and_wine_style(df, "Australia", "Barossa","Australian Cabernet - Shiraz")
aus_barossa_shiraz_data_frame.head()
```
